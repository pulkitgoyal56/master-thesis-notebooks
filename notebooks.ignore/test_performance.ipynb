### Preprocess

```python
# Pre-process Images
saved_preprocessed = 0
loaded_preprocessed = 0
saved_processed = 0
for basename in tqdm(os.listdir(dirname:=BASE_DIR / CATEGORY)):
    if (dirname / basename).is_file() and (dirname / basename).suffix=='.png':
        # Convert to *grid-style*
        if not (dirname / preprocessed_path / basename).is_file() or OVERWRITE:
            file = Im(basename, CATEGORY, base_dir=BASE_DIR, convert=CONVERT, invert=INVERT, skeleton=SKELETON)
            if GRID:
                file.register(GRID_WIDTH, GRID_HEIGHT, GRIDCELL_SIZE, threshold_ratio=THRESHOLD_RATIO, render_w_grid=RENDER_W_GRID, render_type=RENDER_TYPE)
            file.save(preprocessed_path, invert=INVERT_GRID, grid_image=GRID)
            image = file.get_image(grid_image=GRID, invert=INVERT_GRID)
            saved_preprocessed += 1
        else:
            loaded_preprocessed += 1
            image = Im(basename, CATEGORY, preprocessed_path, base_dir=BASE_DIR).image

        # Process for CLIP
        if not (filepath:=dirname / processed_path / (basename.split('.')[0] + '.pt')).is_file() or OVERWRITE:
            image = preprocess(image).to(DEVICE)
            filepath.parent.mkdir(parents=True, exist_ok=True)
            torch.save(image, filepath)
            saved_processed += 1
    if saved_preprocessed + loaded_preprocessed == NUM_IMAGES:
        break
logging.debug(f">> Saved {saved_preprocessed} and loaded {loaded_preprocessed} pre-processed images @ file://{(dirname / preprocessed_path).resolve()}.")
logging.debug(f">> Saved {saved_processed} and loaded {saved_preprocessed + loaded_preprocessed - saved_processed} processed images @ file://{(dirname / processed_path).resolve()}.")
```

### Inference

```python
if MODEL_NAME == 'clip':
    # Pre-process Labels
    logging.debug('> Pre-processing labels...')
    labels = [f'{LABEL_PREFIX}{label}{LABEL_SUFFIX}' for label in categories]
    text = Model.tokenize(labels).to(DEVICE)
    data = lambda image: (image.to(DEVICE), text.to(DEVICE))
    # image_features = model.encode_image(image)
    # text_features = model.encode_text(text)
else:
    data = lambda image: (image.to(DEVICE),)
```

```python
if not (resultpath:=BASE_DIR / CATEGORY / results_path).is_file() or OVERWRITE:
    results = dict()
    images = {}
    logging.info(f'> Running inference on {MODEL_NAME}={MODEL_VERSION} model ...')
    for basename in tqdm(os.listdir(dirname:=BASE_DIR / CATEGORY / processed_path)):
        if (filepath:=dirname / basename).is_file() and filepath.suffix=='.pt':
            images[basename.split('.')[0]] = torch.load(filepath, map_location=DEVICE)

    filenames = list(images.keys())
    images = torch.cat(list(images.values())).view(-1, *images[basename.split('.')[0]].shape)
    with torch.no_grad():
        logits_per_image, *_ = model(*data(images))  # logits_per_text
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

    results = dict(zip(filenames, probs))

    resultpath.parent.mkdir(parents=True, exist_ok=True)
    with open(resultpath, 'wb') as resultfile:
        pickle.dump(results, resultfile)
        logging.debug(f">> Saved results @ file://{resultpath.resolve()}.")
else:
    with open(resultpath, 'rb') as resultfile:
        results = pickle.load(resultfile)
        logging.debug(f">> Loaded results @ file://{resultpath.resolve()}.")
```

#### Results

```python
# sort results in descending order of confidence in first category
results = dict(sorted(results.items(), key=lambda r: r[1][0], reverse=True))
```

```python
results_keys = list(results.keys())
results_values = np.array(list(results.values()))
```

```python
mask = dict(map(lambda filename: (filename, np.argmax(results[filename]) == 0), results))
```

## Graphs

```python
fig = plt.figure(figsize=(15 * (1 + SHOW_IMAGES), 16))
grid = plt.GridSpec(int(np.ceil(len(results_keys) / 10)), 10 * (1 + SHOW_IMAGES), fig)

ax1 = plt.subplot(grid[:, :10])
bottom = 0
for i in range(len(categories)):
    ax1.bar(results_keys, results_values[:, i], 1, bottom=bottom, label=categories[i], color=colors[i])
    bottom += results_values[:, i]
ax1.tick_params(labelbottom=False, labelleft=False)
ax1.axhline(1/len(categories), c='k', lw=0.2)
ax1.axhline(0.5, c='k', lw=0.2)
ax1.margins(0, 0)
ax1.set_xticks([])
ax1.set_yticks([])
handles, labels = ax1.get_legend_handles_labels()
legend = ax1.legend(reversed(handles), reversed(labels), loc='lower left', fontsize='xx-large', ncols=np.ceil(len(categories) / 12))

if SHOW_IMAGES:
    axs = np.array([[plt.subplot(grid[i, 10 + j]) for j in range(10)] for i in range(int(np.ceil(len(results_keys) / 10)))])
    Im.show_images(results_keys, CATEGORY, preprocessed_path, base_dir=BASE_DIR, probabilities=results, invert_grid=INVERT_GRID, mask=mask, axes=axs)

fig.suptitle(TITLE)

plt.show()
```
